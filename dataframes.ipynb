## Q1) Copy the code from the lesson to create a dataframe full of student grades


```python
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})


```

```python
df
```

## a) Create a column named passing_english that indicates whether each student has a passing grade in english.


```python
# (it converts into pandas series,so we can use vectorized operation

df['passing_english']= df['english'] >=70  
df.head()
```

# b) Sort the english grades by the passing_english column. How are duplicates handled?


```python
# by default ascending= True
df.sort_values('passing_english')
```


## c)Sort the english grades first by passing_english and then by student name.

All the students that are failing english should be first, and within the students 
that are failing english they should be ordered alphabetically. 

The same should be true for the students passing english. 

(Hint: you can pass a list to the .sort_values method)
 

```python
df.sort_values(['passing_english','name'])
```

# d) Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.


```python
df.sort_values(['passing_english','english'])
```

## e) Calculate each students overall grade and add it as a column on the dataframe. 

## The overall grade is the average of the math, english, and reading grades.



```python
df['Overall Grade'] = round(df[['math', 'english', 'reading']].mean(axis=1).astype(float))
df
```

```python
df['Overall Grade'].mean() # here axis=o i.e. row-wise mean i.e. allover 
```

## Do any cars have better city mileage than highway mileage?


```python
(df.city >df.highway).sum()
```

```python
# another way
(df.city>df.highway).any()
```

 ## Create a column named mileage_difference this column should contain the difference between highway and city mileage for each car.

```python
df['mileage_difference']=df.highway-df.city
```

```python
df.head(5)
```

## Which car (or cars) has the highest mileage difference?


```python
df['mileage_difference'].max()
```

```python
df.sort_values(by='mileage_difference',ascending=False)
```

```python
df.sort_values(by='mileage_difference').head(2)
```

## Which compact class car has the lowest highway mileage? The best?


```python
# 1st way : highest highway mileage

df[(df['highway']==df['highway'].max() ) & (df['class']=='compact')]
```

```python
# 2nd way : highest highway mileage

df[df['class']=='compact'].sort_values(by='highway', ascending=False).head(1)
```

```python
# 1st way : lowest highway mileage

df[(df['highway']==df['highway'].min() ) & (df['class']=='compact')]
```

```python
# 2nd way : lowest highway mileage

df[df['class']=='compact'].sort_values(by='highway', ascending=False).tail(1)
```

```python
# 3rd way (boolean masks):
df.head(5)
```

```python
df['class'].value_counts()
```

```python
#boolean masks
df['class'] == 'compact'
```

```python
#boolean masks will filter out values 
df [df['class'] == 'compact'].head()
```

```python
# lowest
df [df['class'] == 'compact'].sort_values('highway').head(1) 
```

```python
# highest
df [df['class'] == 'compact'].sort_values('highway').tail(1)
```

## Create a column named average_mileage that is the mean of the city and highway mileage


```python

df['average_mileage']=df[['city', 'highway']].mean(axis=1)
df.head(5)
```

## Which dodge car has the best average mileage? The worst?


```python
# lowest average mileage


df[df['manufacturer']=='dodge'].sort_values(by='average_mileage').head(1)
```

```python
# highest average mileage

df[df['manufacturer']=='dodge'].sort_values(by='average_mileage',ascending=True).tail(1)
```
