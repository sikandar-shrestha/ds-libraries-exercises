## Questions for 'chipotle' database from SQL

Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.



```python
import env
import pandas as pd
import seaborn as sns
```

```python
def get_db_url(hostname, username, password,database):
        return f"mysql+pymysql://{username}:{password}@{hostname}/{database}"

    
```

```python
url=get_db_url(env.hostname,env.username,env.password,'chipotle')
```

```python
df5=pd.read_sql('select * from orders',url)
df5.head()
```

```python
df5.item_price = df5.item_price.str.strip('$').astype(float)
```

```python
df5.head()
```

```python
items_df = df5.groupby('item_name')[['quantity','item_price']].sum()
items_df
```

```python
top_items_df=items_df.sort_values('quantity',ascending=False).head(4)
top_items_df
```

```python
top_items_df.reset_index()
```

```python
sns.barplot(data=top_items_df, y=top_items_df.index, x='item_price');
```
