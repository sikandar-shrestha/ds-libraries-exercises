```python
# Q5.Join the employees and titles DataFrames together. 

url=get_db_url(env.hostname,env.username,env.password,'employees')
```

```python
df_emp=pd.read_sql('select * from employees', url)
df_emp.head()
```

```python
df_title=pd.read_sql('select * from titles', url)
df_title.head()
```

```python
pd.merge(df_emp,df_title,how='outer', right_on='emp_no',left_on='emp_no',indicator=True)
```

```python
df_emp_tit=pd.merge(df_emp,df_title,how='inner', right_on='emp_no',left_on='emp_no')
df_emp_tit
```

```python
df_emp_tit.info()
```

```python
# Q6.For each title, find the hire date of the employee that was hired most recently with that title.


# Convert the 'hire_date' column to a datetime data type
df_emp_tit['hire_date'] = pd.to_datetime(df_emp_tit['hire_date'])
df_emp_tit.head()
```

```python
df_emp_tit.info()
```

```python
# 1st way:
df_emp_tit.groupby('title').max().sort_values('hire_date',ascending=False)
```

```python
# 2nd way:-
# Sort the DataFrame by 'hire_date' in descending order within each 'title' group
df_sorted=df_emp_tit.sort_values(by=['title', 'hire_date'], ascending=[True, False])
df_sorted
```

```python
# Get the most recent hire date for each title
df_sorted.groupby('title')['hire_date'].first()
```

```python
# Q7..Write the code necessary to create a cross tabulation of the number of titles by department. 
# (Hint: this will involve a combination of SQL code to pull the necessary data and python/pandas code 
# to perform the manipulations.)

df_depts=pd.read_sql('''
select *
from departments
join dept_emp
    using(dept_no)
join titles
    using(emp_no)''',url)
df_depts.head()
```

```python
pd.crosstab(df_depts.title,df_depts.dept_name)
```
