## Q2) Create and label 4 separate charts with the following equations (choose a range for x that makes sense):

    - y =âˆšx

    - y = x3

    - y = 2x

    - y =1/(x+1)

You can use functions from the math module to help implement some of 

the equations above.

```python
import math as m
import matplotlib.pyplot as plt
```

```python
x=range(50)
```

```python
# we can't apply vectorization operator, since x is in the form of list

y= [m.sqrt(i) for i in x ] 
plt.plot(x,y)
plt.title(" plot of y=âˆšx")
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
y= [i**3 for i in x]

plt.plot(x,y)
plt.title(" plot of y=x^3")
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
y= [2*i for i in x ]

plt.plot(x,y)
plt.title(" plot of y=2x")
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
y= [1/(i+1) for i in x ]

plt.plot(x,y)
plt.title(" plot of y =1/(i+1)")
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

## Q3) Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# 1st way (/w converting array )

plt.figure(figsize=(12,6))

plt.subplot(221) #subplot(#ofrows, #ofcolumns,#position of plot)
x = np.array(range(0,11))
y = x**0.5
plt.plot(x,y)

plt.subplot(222)
x = np.array(range(-10,11))
y = x**3
plt.plot(x,y)

plt.subplot(223)
x = np.array(range(0,11))
y = 2**x
plt.plot(x,y)

plt.subplot(224)
x = np.array(range(-10,10))
y = 1 / (x+1)
plt.plot(x,y)

# plt.title('test title')
plt.suptitle('four subplots') #title for subplots
plt.tight_layout() #ensures i can see my axises
plt.show()





```

```python
### Another way ( w/o converting array )
x=range(50)
# Define the equations
y1 = [m.sqrt(i) for i in x ]
y2 = [i**3 for i in x]
y3 = [2*i for i in x]
y4 = [1/(i+1) for i in x ]

# Create and label the charts
plt.figure(figsize=(12, 8))  # Set the figure size

# Chart 1: y = âˆšx
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='y = âˆšx')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Chart 2: y = x^3
plt.subplot(2, 2, 2)
plt.plot(x, y2, label='y = x^3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Chart 3: y = 2x
plt.subplot(2, 2, 3)
plt.plot(x, y3, label='y = 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Chart 4: y = 1/(x+1)
plt.subplot(2, 2, 4)
plt.plot(x, y4, label='y = 1/(x+1)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Adjust layout for better visualization
plt.tight_layout()

# Show the plots
plt.show()


```
